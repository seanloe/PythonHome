## Data Cleaning

```python
from mypack.preprocessing.nan_process import *
schools_cpy = schools.copy()
#first 3 has very high percentage of NaN 
#fill with 0 for later 1-hot
schools[ds[:3].index] = schools[ds[:3].index].fillna(0)
# 3rd is income which i choose to fill with mean
schools['School Income Estimate'] = (
    schools['School Income Estimate']
    .apply(lambda x:convert_finan_digits(x))
    .astype('float32')
    #.fillna(schools['School Income Estimate'].mean())
)
schools['School Income Estimate'] = schools['School Income Estimate'].fillna(schools['School Income Estimate'].mean())
#schools['School Income Estimate']
#next 6 cols catergorical
schools[ds[4:11].index] = schools[ds[4:11].index].fillna('empty', axis = 1)
#schools['Effective School Leadership Rating']
schools['Average Math Proficiency'] = schools['Average Math Proficiency'].fillna(schools['Average Math Proficiency'].mean())
schools['Average ELA Proficiency'] = schools['Average ELA Proficiency'].fillna(schools['Average ELA Proficiency'].mean())

schools[ds[13:21].index] = (
    schools[ds[13:21].index]
    .applymap(lambda x:convert_finan_digits(x))
    .astype('float32')
    )
schools['Economic Need Index'] = schools['Economic Need Index'].fillna(schools['Economic Need Index'].mean())
schools.head(20)
```
