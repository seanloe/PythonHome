# Movan Machine learning. Keras learning

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Activation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generate Data

```python
X = np.linspace(-1,1,200)
Y = 0.5*X+2+np.random.normal(-0.1,0.1,200)
#plt.scatter(X,Y, color='b')
ll = list(zip(X,Y))

df = pd.DataFrame(list(zip(X,Y)), columns=['X','Y'])

#train test split
X_train, X_test = X[:160], X[160:]
Y_train, Y_test = Y[:160], Y[160:]
sns.lmplot(x='X', y='Y', data=df, fit_reg=False)
```

## Now build the model

```python
model = Sequential()
model.add(Dense(units =1, input_dim=1))
model.add(Activation('relu'))
#define loss function and optimizer
model.compile(loss='mse', optimizer='sgd')

```

## Training

```python
loss_list = model.fit(X_train, Y_train, batch_size=20, epochs=200)
score = model.evaluate(X_test, Y_test, batch_size=20)
#print(score)
W,b = model.layers[0].get_weights()
print('Weight:%.2f,Bias:%.2f'%(W,b))
#print(loss_list.history)
```

# Ok let's do a more complicate linear regression

## Generate Data

```python
SAMPLE_NUM = 5000
X2 = np.linspace(-np.pi,np.pi,SAMPLE_NUM)
Y2 = 3-2*np.cos(X2)**2+np.random.normal(-0.1,0.1,SAMPLE_NUM)
#plt.scatter(X,Y, color='b')
ll = list(zip(X2,Y2))

df2 = pd.DataFrame(list(zip(X2,Y2)), columns=['X','Y'])
df2 = df2.reindex(np.random.permutation(df2.index))

#train test split
#X2_train, X2_test = X2[:int(SAMPLE_NUM*0.8)], X2[int(SAMPLE_NUM*0.8):]
#Y2_train, Y2_test = Y2[:int(SAMPLE_NUM*0.8)], Y2[int(SAMPLE_NUM*0.8):]
from sklearn.model_selection import train_test_split
X2_train, X2_test, Y2_train, Y2_test = train_test_split(df2.X,df2.Y)
sns.lmplot(x='X', y='Y', data=df2, fit_reg=False)
```

```python
model2 = Sequential()
model2.add(Dense(10, input_dim=1))
model2.add(Activation('relu'))
model2.add(Dense(output_dim=5))
model2.add(Activation('relu'))
model2.add(Dense(output_dim=1))
#model2.add(Activation('relu'))
model2.compile(loss='mse',optimizer='sgd')

model2.fit(X2_train,Y2_train,batch_size=100, epochs=100)
model2.evaluate(X2_test, Y2_test)
```

## visualize current trainer model

```python
X_space = np.linspace(-np.pi,np.pi,50)
result = model2.predict(X_space)
#print(result)
plt.plot(X_space,result,color='r')
```
